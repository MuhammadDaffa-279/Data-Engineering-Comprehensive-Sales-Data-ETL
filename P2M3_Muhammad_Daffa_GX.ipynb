# Creating Data Context

In this section, we will be performing the following:
1. Creating Data Context
2. Initialize datasource
3. Creating a name for the data asset
4. Build the batch request
5. Creating the Expectation Suite
6. Create the Validator

```python
# Creating Data Context and root to this file
context = FileDataContext.create(project_root_dir='./')
```

After creating the data context, we would initialize the datasource, defining data asset name and building the batch request

```python
# Initialize Datasource
datasource_name = 'Superstore Milestone 3'

# Delete the datasource (in the context of restart run all)
if datasource_name in context.datasources:
  context.sources.delete(datasource_name)

datasource = context.sources.add_pandas(datasource_name)

# Create name for the data asset
asset_name = 'Sales Milestone 3'
path_to_data = '/content/P2M3_Muhammad_Daffa_Data_Clean.csv'
asset = datasource.add_csv_asset(asset_name, filepath_or_buffer=path_to_data)

# Build batch request
batch_request = asset.build_batch_request()
```

We have successfully built the batch request. Next, we would like to create an expectation suite and the validator

```python
# Create an expectation suite
expectation_suite_name = 'Project Milestone 3 Daffa'
context.add_or_update_expectation_suite(expectation_suite_name)

# Create a validator using above expectation suite
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = expectation_suite_name
)

# Check the validator
validator.head()
```

We have successfully create a validaton. Then, our data is ready to be validated with some expectation.

---

# Expectation

This section will perform the data validation with 7 expectations

First thing, we need to make sure that our "row_id" is unique, since this row is the primary key for our dataset. We dont want any duplicated value inside of our dataset.

```python
# Validation 1 - "row_id" as the primary key To be Unique

validator.expect_column_values_to_be_unique('row_id')
```

First validation is success. We have validated that there are no duplicated value in our dataset, since the primary key unique validation is passed.

Since we are going to address and evaluate the sales in the dataset. We need to make sure that the column "sales" contain non missing values.

```python
# Validation 2 - "sales" column Non-Null
validator.expect_column_values_to_not_be_null('sales')
```

The second validation to make sure there are no missing value inside the sales column is success.

For the next validation, we also need to make sure that the value of the sales does not contain negative value and in between reasonable sales range.

```python
# Validation 3 - "sales" column To be between 0 (non negative) up to 100000
validator.expect_column_values_to_be_between('sales', min_value=0, max_value = 100000)
```

The third validation is success. We have a non negative sales and within reasonable range inside of our dataset. Thus, it can be concluded that our sales column in the dataset is realiable for further processing.

Further, we also want to make sure that inside the dataset, we are selling the product of our store (consistency). Inside of our store, we are selling furniture, office supplies and technology category.

```python
# Validation 4 - To be in set (Category product)
validator.expect_column_values_to_be_in_set('category', ['Furniture', 'Office Supplies', 'Technology'])
```

We have successfully validated that inside the dataset, we are consistent in selling the product that is available in our store.

We already have checked the sales column that it does not contain missing value and the value is greater than 0. We want to make sure that the data type in the column is in float type just to double check the sales column.

```python
# Validation 5 - "sales" to be in float type
validator.expect_column_values_to_be_of_type('sales', type_='float')
```

The sales column is in float type. Therefore, the sales column is validated completely and comprehensively.

Moving on, we also want to check that the product id column is available in our dataset. We want to know which specific items we have sell to our customer

```python
# Validation 6 - Column "product_id" to be existed
validator.expect_column_to_exist('product_id')
```

Validating to the product_id column is success. We know that our dataset is already consist of the product id column (item sold).

For the last validation, we want to make sure that we have all the informations/columns for out sales table. Our standard sales table consist of 18 columns. Thus, we need to validate that the dataset already comply with this standard.

```python
# Validation 7 - Table column count to be equal to 18 (our standard dataset for sales)
validator.expect_table_column_count_to_equal(18)
```

Last validation is success, we have all the informations required in the sales data.

In conclusion, we have conducted 7 essential validations process to make sure our data is reliable and ready for further processing.
